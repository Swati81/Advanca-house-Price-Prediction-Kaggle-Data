## removing few outliers

```python
def Z_score(data,left,right):
    index= []
    mean = data.mean()
    std = data.std()
    for i in range(len(data)):
        z = (data.iloc[i]-mean)/std
        if (z >= -left) and (z <= right):
            index.append(i)
        else:
            pass
    return index 
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score,mean_squared_error
#rfr = RandomForestRegressor(random_state=0)
xgr = xgb.XGBRegressor(objective='reg:linear',random_state=0)
```

```python
def modelTrain(clf,x,y,rs=0):
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.34,random_state=rs)
    clf.fit(xtrain,ytrain)
    pred = clf.predict(xtest)
    accu = r2_score(ytest,pred)
    error = np.sqrt(mean_squared_error(ytest,pred))
    print(f'Features : {xtrain.shape[1]} | R2-Score :{round(accu*100,2)} % | RMSE : {round(error)}')
```
